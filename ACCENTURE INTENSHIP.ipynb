The content data frame looks clean enough to be merged with the other two data frame. Before i will proceed to do this, i will preview all the three different data frames before merging them together for further analysis

```python
reactions = reactions.dropna(subset=['Type'], axis=0)
reactions.head()
```

```python
#merge tables on content ID columns
Performance = reactions.merge(content,on='Content ID')
Performance
```

```python
#merge tables on  type column
Buzz_Performance = Performance.merge(reaction_types,on='Type')
Buzz_Performance
```

```python
# I will further like to change the column "Type" to "reaction type " to make the column more meaningfull

Buzz_Performance = Buzz_Performance.rename(columns={'Type': 'reaction type'})
Buzz_Performance
```

```python
#group by category and calculate total score
popular_five = Buzz_Performance.groupby('Category').sum().sort_values(by='Score',ascending=False)
popular_five.head()
```

```python
# writing the result into a csv file
popular_five.to_csv('social_buzz_popular_five.csv')
```

```python
# writing the result into an excel file
popular_five.to_excel('social_buzz_popular_five.xlsx')
```
